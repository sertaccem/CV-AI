```python
from PIL import Image
import os
import seaborn as sns
import matplotlib.pyplot as plt

# Function to get image sizes
def get_image_sizes(directory):
    image_sizes = []
    for class_name in ['NORMAL', 'PNEUMONIA']:  # Iterate through each class
        class_dir = os.path.join(directory, class_name)
        for image_name in os.listdir(class_dir):  # Iterate through each image in the class
            image_path = os.path.join(class_dir, image_name)
            
            # Filter out non-image files
            if image_name.endswith(('.png', '.jpg', '.jpeg')):  # Only process images
                with Image.open(image_path) as img:  # Open the image using PIL
                    width, height = img.size  # Get the width and height
                    image_sizes.append((width, height))  # Store the size as a tuple
    return image_sizes

# Get the image sizes from the train directory
train_image_sizes = get_image_sizes(train_directory)

# Separate the sizes into width and height lists
image_widths = [size[0] for size in train_image_sizes]
image_heights = [size[1] for size in train_image_sizes]

# Plot the distribution of image widths
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(image_widths, bins=20, kde=True, color='blue')
plt.title('Distribution of Image Widths (Train Set)')
plt.xlabel('Width (pixels)')
plt.ylabel('Frequency')

# Plot the distribution of image heights
plt.subplot(1, 2, 2)
sns.histplot(image_heights, bins=20, kde=True, color='green')
plt.title('Distribution of Image Heights (Train Set)')
plt.xlabel('Height (pixels)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
